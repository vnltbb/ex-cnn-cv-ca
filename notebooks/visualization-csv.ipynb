```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 업로드된 파일들 다시 읽기
file_paths = {
    "EfficientNetB0": "/Users/vnlt/PROJECT/ex-cnn-cv/utils/csv-results/EfficientNetB0.csv",
    "DenseNet121": "/Users/vnlt/PROJECT/ex-cnn-cv/utils/csv-results/DenseNet121.csv",
    "ResNet50": "/Users/vnlt/PROJECT/ex-cnn-cv/utils/csv-results/ResNet50.csv"
}
models = {
    'EfficientNetB0': file_paths['EfficientNetB0'],
    'ResNet50': file_paths['ResNet50'],
    'DenseNet121': file_paths['DenseNet121']
}
```

```python
for model_name, file_path in models.items():
    df = pd.read_csv(file_path)
    plt.plot(df['epoch'], df['val_accuracy'], label=f'{model_name} Val')
    plt.plot(df['epoch'], df['accuracy'], label=f'{model_name} Train')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Train vs Val Accuracy per Model')
plt.legend()
plt.grid(True)
plt.show()
```
